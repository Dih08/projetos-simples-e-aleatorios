```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('netflix_titles_nov_2019[1].csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
#Olhando quantos valores nulos tem
df.isnull().sum()
```

```python
#tirando os valores null
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(10,10))
sns.countplot(df['type'])
```

```python
df.head()
```

```python
#Checando valores unicos
df.nunique()
```

```python
#checando valores duplicados
df.duplicated().sum()
```

```python
import datetime as dt
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
df['year_added'].astype(int)
df['month_added'].astype(int)
```

```python
df.head()
```

```python
#vamos ver as classificações entre tipo e idade (classificação)
df.groupby('rating')['type'].value_counts()
```

```python
#Vamos representar isso ne um grafico
plt.figure(figsize=(10,10))
sns.countplot(x='rating', hue='type', data=df)
plt.show()
```

```python
#agora no rafico de pizza
plt.figure(figsize=(10,10))
df.groupby('rating')['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```
