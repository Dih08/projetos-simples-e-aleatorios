# Teorema Navie Bayes

## Importando os dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importando os dados

```python
dados = pd.read_csv('Social_Network_Ads.csv')
X = dados.iloc[:, :-1].values
y = dados.iloc[: , -1].values
```

```python
print(X)
```

```python
print(y)
```

## Dividindo os dados por treino e teste

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y ,test_size= 0.25,random_state=0 )
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Escalando os recursos

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Treinando Navie Bayes em todo o modelo

```python
from sklearn.naive_bayes import GaussianNB
classificador = GaussianNB()
classificador.fit(X_train, y_train)
```

### Valores predict novo resultados

```python
print(classificador.predict(sc.transform([[30,87000]])))
```

### Teste predict

```python
y_pred = classificador.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Usando a matriz de confuso

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Visualizando resultado de treino

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classificador.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Naive Bayes (Treino)')
plt.xlabel('Idade')
plt.ylabel('Salario Estimado')
plt.legend()
plt.show()
```
