```python
import urllib.request as urllib2
from bs4 import BeautifulSoup
```

```python
pagina = urllib2.urlopen("https://www.wikipedia.org/wiki/Brasil")
html_doc = pagina.read()
```

```python
#parseando  e arruamndo a pagina
soup = BeautifulSoup(html_doc, 'html.parser')
#soup formatado
formatado = soup.prettify()
```

```python
type(soup)
```

```python
#navegando pelas tags
soup.body.p
```

```python
paragrafos = soup.find_all('p')
```

```python
type(paragrafos)
```

```python
texto = soup.get_text()
print(texto)
```
