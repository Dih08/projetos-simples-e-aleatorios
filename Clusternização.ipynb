# K-Means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

from sklearn.cluster import KMeans
```

```python
X, _ = make_blobs(n_samples=300 , centers= [[0,0], [1,0], [0.75, 1.25]], cluster_std = 0.2 , random_state = 0)
X.shape
```

```python
X
```

```python
plt.figure(figsize=[8,6], dpi=100)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
km = KMeans(n_clusters=3)
km
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```

```python
plt.figure(figsize=[8,6], dpi=100)
plt.scatter(X[:,0], X[:,1])
plt.title('centroides', fontsize=14 , fontweight= 'bold')
for centr in km.cluster_centers_:
    plt.scatter(centr[0], centr[1], marker='x', s=300, c='red',lw = 7)
plt.show()
```

```python
km.predict(X)
```

```python
plt.figure(figsize=[8,6], dpi=100)
plt.scatter(X[:,0], X[:,1], c = km.predict(X), cmap = plt.cm.Paired)
plt.title('clusternizao dos dados:', fontsize=14 , fontweight= 'bold')
for centr in km.cluster_centers_:
    plt.scatter(centr[0], centr[1], marker='x', s=300, c='black',lw = 7)
plt.show()
```

```python
plt.figure(figsize=[20,50], dpi=100)

for k in range(1,13):
    km = KMeans(n_clusters=k)
    km.fit(X)
    plt.subplot(6,2,k)
    plt.title('Qtd Clusters' + str(k), fontsize = 14, fontweight = 'bold')
    plt.scatter(X[:,0], X[:,1], c = km.predict(X), cmap = plt.cm.Paired)
    for centr in km.cluster_centers_:
        plt.scatter(centr[0], centr[1], marker='x', s=300, c='black',lw = 7)
    plt.xticks([])
    plt.yticks([])
plt.show()
```


A frmula matemtica que representa o WCSS (Within-Cluster Sum of Square)  a soma das distncias ao quadrado de cada ponto de dados ( x_i ) at o centroide do cluster ( C_k ) ao qual ele pertence. Matematicamente, pode ser expressa como:





=


=
1















2
WCSS=k=1KiCkxik2

Onde:

( K )  o nmero de clusters.
( C_k )  o conjunto de pontos no cluster ( k ).
( x_i )  um ponto de dados dentro do cluster ( C_k ).
( \mu_k )  o centroide do cluster ( k ).
( ||x_i - \mu_k|| )  a distncia euclidiana entre o ponto de dados ( x_i ) e o centroide ( \mu_k ).

```python
#Curva cotovelo

lista_inercia = []
for i in range(1,13):
    km = KMeans(n_clusters=i)
    km.fit(X)
    lista_inercia.append(km.inertia_)

    # e vamos plotar o cotovelo

plt.figure(figsize=[15,6])
plt.title('Curva Cotovelo', fontsize=14, fontweight='bold')
plt.xlabel('Qtd de Clusters', fontsize=14 , fontweight = 'bold')
plt.ylabel( 'WCSS', fontsize=14 , fontweight = 'bold')
plt.plot(range(1, 1+len(lista_inercia)), lista_inercia, c = 'black',lw = 2)
plt.scatter(range(1, 1+len(lista_inercia)), lista_inercia, c = 'black',lw = 2)
plt.xticks(range(1, 1+len(lista_inercia)))
plt.grid()
plt.show()
```

# Case pratico:

```python
df = pd.read_csv('countries_of_the_world[1].csv').dropna().reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
dff = df[['Country', 'Infant mortality (per 1000 births)', 'GDP ($ per capita)']].copy()
dff.columns = ['pas','mortalidade_infantil', 'renda_per_capta']
dff.head()
```

```python
dff.mortalidade_infantil = dff.mortalidade_infantil.apply(lambda s: str(s).replace(',', '.')).astype(float)
```

```python
dff
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
Escala = sc.fit_transform(dff.values[:, 1:])
```

```python
plt.figure(figsize=[8,6])
plt.scatter(Escala[:,0], Escala[:,1])
plt.xlabel('Mortalidade:' , fontsize=14 , fontweight = 'bold')
plt.ylabel('Renda:' , fontsize=14 , fontweight = 'bold')
plt.grid()
plt.show()
```

```python
# Excluindo colunas no numricas do DataFrame
dff_numeric = dff.select_dtypes(include=[np.number])

# Calculando a correlao apenas com as colunas numricas
correlacao = dff_numeric.corr()
print(correlacao)
```

```python
L=[]
for k in range(1, 21):
    km = KMeans(n_clusters=k)
    km.fit(Escala)
    L.append(km.inertia_)



plt.figure(figsize=[15,6])
plt.title('curva de Eibow', fontsize = 20 , fontweight = 'bold')
plt.xlabel('Qtd de clusters', fontsize = 20 , fontweight = 'bold')
plt.ylabel('WCSS:', fontsize = 20 , fontweight = 'bold')
plt.plot(range(1, 1+len(L)), L, c = 'black', lw=2) # Corrected this line
plt.scatter(range(1, 1+len(L)), L, c='black', lw=2)
plt.xticks(range(1, 1+len(L)))
plt.grid()
plt.show()
```
