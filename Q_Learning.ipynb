```python
#bibliotecas

import numpy as np
import pandas as pd
```

```python
#definindo estado e ações a sera, tomadas

states = [0,1,2,3,4,5]
actions = [0,1,2,3,4,5]

#parametro gamma

gamma = 0.8
goal_state = 5
```

```python
#matriz de recompensa
R = [[-1,-1,-1,-1,0,-1],
     [-1,-1,-1,0,-1,100],
     [-1,-1,-1,0,-1,-1],
     [-1,0,0,-1,0,-1],
     [0,-1,-1,0,-1,100],
     [-1,0,-1,-1,0,100]]

R = np.array(R)
print('Matriz de recompensa')
print(pd.DataFrame(R, index=['state_'+str(i) for i in range(R.shape[0])],
                   columns=['state_' +str(i) for i in range(R.shape[0])],))
```

```python
#Inicializando a matriz 0 -zerada
Q = np.zeros((len(states), len(actions)))
print('Q-matrix')
print(pd.DataFrame(Q, index=['state_'+str(i) for i in range(Q.shape[0])],
                   columns=['state_' +str(i) for i in range(Q.shape[0])],))
```

```python
#função auxiliar
def get_possible_next_state(R,state):
  possible_next_states = np.argwhere(R[state,: ] >= 0).reshape(-1,)
  return possible_next_states
```

```python
#examplos
get_possible_next_state(R, state = 0)
```

```python
get_possible_next_state(R, state = 3)
```

```python
#construindo um episódio


#escolhendo o estado
state = np.random.choice(states, size = 1)[0]

#looping while
next_state = None
while next_state != goal_state:

#escolhendo os proximos estados
   possible_next_states = get_possible_next_state(R, state)

#Definimos o proximo estado
   next_state = np.random.choice(possible_next_states, size = 1)[0]
   print('estad0 --> proximo estado ',state,next_state)

#  Atuslizando de acordo com as decisões tomadas
   M = Q[next_state, get_possible_next_state(R, next_state)].max()
   Q[state, next_state] = R[state, next_state] + gamma * M

   state = next_state
   print()

print('fim')
print()
print('q-matrix atualizado')
print(Q)
```

```python
class Q_learning:

  def __init__(self, states, actions,R, gamma, goal_state):
    self.states = states
    self.actions = actions
    self.R = R
    self.gamma = gamma
    self.goal_state = goal_state
    self.Q = np.zeros(len(states)* len(actions)).reshape(len(states), len(actions))
    self.Q = Q

  def get_Qmatrix(self):
    Qdf = pd.DataFrame(self.Q, index=['state_'+str(i) for i in range(self.Q.shape[0])],
                   columns=['action_' +str(i) for i in range(self.Q.shape[1])]).astype(int)
    return Qdf

  def run_episode(self):
      state = np.random.choice(self.states, size = 1)[0]
      next_state = None
      while next_state != self.goal_state:
         possible_next_states = get_possible_next_state(R, state)
         next_state = np.random.choice(possible_next_states, size = 1)[0]
         M = self.Q[next_state, get_possible_next_state(R, next_state)].max()
         self.Q[state, next_state] = self.R[state, next_state] + self.gamma * M
         state = next_state

  def train_agent(self, n_episodes):
     for e in range(n_episodes):
       self.run_episode()
```

```python
#instanciando um agente
agente = Q_learning(states=states,
                   actions=actions,
                   R=R,
                   gamma=gamma,
                   goal_state=goal_state)
```

```python
agente.get_Qmatrix()
```

```python
agente.train_agent(n_episodes=10)
agente.get_Qmatrix()
```
